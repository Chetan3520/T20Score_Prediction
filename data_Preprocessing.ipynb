```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
```

## Accessing all the file name from YAML files

```python
filenames = []
for file in os.listdir('data'):
    filenames.append(os.path.join('data',file))
    
#code Explanation: Go to the file and list out all names of YAML files and append it to filenames
```

```python
filenames[0:5]
```

```python
final_df = pd.DataFrame()
counter = 1
for file in tqdm(filenames):
    with open(file, 'r') as f:
        df = pd.json_normalize(safe_load(f))
        df['match_id'] = counter
        final_df = pd.concat([final_df, df], ignore_index=True)
        counter += 1    
final_df
```

###### Explanation of the above code step by steps

__final_df = pd.DataFrame():__ This line creates an empty table called final_df to store the data.

__counter = 1__ This line initializes a counter variable with a value of 1. It will be used to assign a unique match ID to each JSON file.

__for file in tqdm(filenames)::__ This loop iterates over a list of file names. The tqdm function provides a progress bar to track the loop's execution.

__with open(file, 'r') as f::__ This line opens the current file in read mode and assigns it to the variable f. The with statement ensures that the file is closed properly after reading.

__df = pd.json_normalize(safe_load(f)):__ This line reads the contents of the file, converts the JSON data into a tabular format using pd.json_normalize(), and creates a DataFrame called df.

__df['match_id'] = counter__ This line adds a new column named 'match_id' to the DataFrame df and assigns the current value of the counter to all rows in that column.

__final_df = pd.concat([final_df, df], ignore_index=True):__ This line concatenates the current DataFrame df with the final_df DataFrame, combining the data from all files. The resulting DataFrame is assigned back to final_df.

__counter += 1:__ This line increments the counter by 1 to prepare it for the next file. This ensures that each file gets a unique match ID.

__Finally__, final_df contains the consolidated data from all the JSON files. It is the result of combining and merging the individual DataFrames into a single DataFrame.

## extract meaningfull data (all record corresponding to individual ball)

```python
count = 1
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_of_dismissed,
            'city':city,
            'venue':venue
        })
    delivery_df = pd.concat([delivery_df, loop_df], ignore_index=True)
```

###### explanation of the above code

The code is performing data extraction and transformation using pandas library in Python.
Here's a breakdown of the code:

__count = 1:__ This line initializes a counter variable with a value of 1.

__delivery_df = pd.DataFrame():__ This line creates an empty DataFrame called delivery_df that will store the extracted data.

__for index, row in matches.iterrows()::__ This loop iterates over each row in the DataFrame called matches. It uses the __iterrows()__ method to access both the index and row data.

The first __if__ statement checks if the current value of count matches any of the numbers in the list __[75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345].__ If it does, __count__ is incremented by 1 and the loop continues to the next iteration using __continue.__ This condition appears to exclude certain matches from processing.

__ball_of_match, batsman, bowler, runs, player_of_dismissed, teams, batting_team, match_id, city, and venue__ are all empty lists created to store specific data fields extracted from each row.

The nested loop begins with __for ball in row['innings'][0]['1st innings']['deliveries']:__, which iterates over the deliveries in the first innings of each row.

The inner loop __for key in ball.keys():__ iterates over each key in the current delivery.

Inside the inner loop, various fields such as __match_id, batting_team, teams, ball_of_match, batsman, bowler, runs, city, and venue__ are extracted from the nested structure of the __row__ and __ball__ objects and appended to the respective lists.

The extracted data is then used to create a new DataFrame called __loop_df__ using the __pd.DataFrame()__ constructor.

__delivery_df = pd.concat([delivery_df, loop_df], ignore_index=True):__ This line concatenates (joins) the current DataFrame loop_df with the delivery_df DataFrame. The resulting DataFrame is assigned back to __delivery_df__. The __ignore_index=True__ parameter ensures that the concatenated DataFrame has a continuous index, disregarding the original indices of the individual DataFrames.

The loop continues with the next iteration, processing the next row in the __matches__ DataFrame.

Finally, the code produces the __delivery_df__ DataFrame, which contains the extracted data from the specified matches. This DataFrame will have columns such as __match_id, teams, batting_team, ball, batsman, bowler, runs, player_dismissed, city, and venue.__
