## data Extracton

```python
import pandas as pd
import pickle
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pickle.load(open('dataset_level2.pkl','rb'))
```


#### The necessary features corresponding for ML model

```python
df

# batting_team
# bowling team
# city
# current score
# ball left
# wickets left
# current rr
# last five
```

#### check null values

```python
df.isnull().sum()
```

#### how many cities which have null values corresponding to venue?

```python
df[df['city'].isnull()]['venue'].value_counts()
```

#### (city has null values because we can observe that the cities are mention in corresponding venue column so lets extract that cities from venue column)

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df
```

#### Going to apply threshold that will consider city eligible when at least 600 matches played in that city

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)];df
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
#df.groupby('match_id').cumsum()['runs']
```

#### seperate the ball and over values and add the new columns in the data

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

#### Find out the numbers of balls bowled

no of balls bowled= over no. *6+ current ball no

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

#### find out balls left

```python
df['balls_left'] = 120 - df['balls_bowled']
#balls may be more than 120 for that this code will give negative values to avoid this,
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

#### find out how many wickets left

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
df
```

```python
df['wickets_left']=10-df['player_dismissed']
```

```python
df
```

```python
df.head()
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
print(df['runs'].dtype)
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30)['runs'].sum().values.tolist())
```

```python
df['last_five'] = last_five
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.to_excel('T20.xlsx',index=False)
```

```python
final_df.sample(2)
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
X_train.columns
```

```python
#!pip install xgboost==1.4.2
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
